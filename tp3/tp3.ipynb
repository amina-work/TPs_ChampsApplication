```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#listez les données
soildata = pd.read_csv("data.csv")
soildata.head()
```

```python
#q3
soildata = soildata.rename(columns={"moisture": "Humidité", "temp": "Température", "pump":"Arrosage"})
soildata.head()
```

```python
#Q4
soildata.describe()
```

```python
#Q5
value_counts = soildata['Arrosage'].value_counts()
print(value_counts)
```

```python
#Q6
labels = ['arrosée', 'non arrosée']
plt.pie(value_counts, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title("Distribution de l'arrosage")
plt.show()
```

```python
#Q7
plt.scatter(soildata['Humidité'], soildata['Arrosage'], color="blue", marker='+', label='Arrosage')
plt.xlabel('Humidité')
plt.ylabel('Arrosage')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.scatter(soildata['Température'], soildata['Arrosage'], color="blue", marker='+', label='Arrosage')
plt.xlabel('Température')
plt.ylabel('Arrosage')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q9
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(soildata["Humidité"], soildata["Température"], soildata["Arrosage"].map({1: 5, 0: 5}), c=soildata["Arrosage"].map({1: 'blue', 0: 'red'}), linewidths=1, alpha=0.8, edgecolors="k")
ax.set_xlabel('Humidité')
ax.set_ylabel('Température')
ax.set_zlabel('Arrosage')
plt.show()
```

```python
#Q10
new_c = ["Humidité", "Température", "Arrosage"]
df = soildata[new_c]
cor = df.corr()
r_corr = cor.round(2)
sb.heatmap(data=r_corr, annot=False)
plt.show()
```
