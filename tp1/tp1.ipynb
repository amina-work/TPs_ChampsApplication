```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#q1
df = pd.read_csv("IOT-temp.csv")
df.head()
```

```python
#q2
df.describe()
```

```python
#Q3
print("La température maximale: ", max(df["temp"]))
print("La température minimum: ", min(df["temp"]))
print("La température moyenne: ", df["temp"].mean())
print("L'ecart type de la température: ", df["temp"].std())
```

```python
#Q4
pivot_moy = df.pivot_table(values="temp", index='out/in', aggfunc='mean')
pivot_max = df.pivot_table(values="temp", index='out/in', aggfunc='max')
pivot_min = df.pivot_table(values="temp", index='out/in', aggfunc='min')

print("Tableau de moyennes: \n", pivot_moy)
print("Tableau de maximale: \n", pivot_max)
print("Tableau de minimum: \n", pivot_min)
```

```python
#Q5
df_out = df[df['out/in'] == 'Out']
df_in = df[df['out/in'] == 'In']

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 6))

axes[0].plot(df_out.index, df_out['temp'])
axes[1].plot(df_in.index, df_in['temp'])

plt.show()
```
